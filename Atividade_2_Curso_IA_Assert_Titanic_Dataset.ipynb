```python
import pandas as pd
from seaborn import load_dataset
```

#Carregar a Base de Dados


```python
base_dados = load_dataset("titanic")
base_dados
```

```python
NaN_counter = pd.isnull(base_dados["deck"])

base_dados[NaN_counter]
```

```python
base_dados.shape
```

#Tratamento de Dados Indicados pelo Professor na Vídeoaula

```python
base_dados=base_dados.drop(['class','deck'], axis=1) #Professor indicou a remoção das 2 colunas
base_dados
```

```python
base_dados = base_dados.dropna()
```

```python
base_dados
```

#Normalização de Dados e OneHotEncoder


```python
from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()

base_dados[['age']] = mms.fit_transform(base_dados[['age']])
base_dados
```

```python
base_dados[['fare']] = mms.fit_transform(base_dados[['fare']])
base_dados
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer

column_transformer = make_column_transformer((OneHotEncoder(), ['sex','embarked','who','adult_male','embark_town','alive','alone']), remainder = 'passthrough')

base_dados = column_transformer.fit_transform(base_dados)

base_dados = pd.DataFrame(data=base_dados, columns=column_transformer.get_feature_names_out())
```

```python
base_dados
```

#Label Encoder

```python
##Colunas 18(survived) e 19(pclass) são os targets

from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()

base_dados['remainder__pclass'] = label_encoder.fit_transform(base_dados['remainder__pclass'])

#base_dados['remainder__survived'] = label_encoder.fit_transform(base_dados['remainder__survived'])

```

```python
base_dados
```

#Separar Conjuntos de Treino e Teste

```python
y = base_dados['remainder__pclass']

#y = base_dados['remainder__survived']

X = base_dados.drop(['remainder__pclass'], axis=1)

#X = base_dados.drop(['remainder__survived'], axis=1)

```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=None,stratify=y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

model1 = DecisionTreeClassifier(criterion="entropy")
model2 = KNeighborsClassifier(n_neighbors=7, metric='euclidean',algorithm='brute')

model3 = DecisionTreeClassifier(criterion="log_loss", max_depth=8, splitter="best")
model4 = KNeighborsClassifier(n_neighbors=5, metric='minkowski',algorithm='ball_tree')


model1.fit(X_train, y_train)
model2.fit(X_train, y_train)
model3.fit(X_train, y_train)
model4.fit(X_train, y_train)


```

```python
from sklearn import metrics

result1 = model1.predict(X_test)
result2 = model2.predict(X_test)
result3 = model3.predict(X_test)
result4 = model4.predict(X_test)

acc1 = metrics.accuracy_score(result1, y_test)
show1 = round(acc1 * 100)
print("Resultado modelo 1(em %): ", format(show1))

acc2 = metrics.accuracy_score(result2, y_test)
show2 = round(acc2 * 100)
print("Resultado modelo 2(em %): ", format(show2))

acc3 = metrics.accuracy_score(result3, y_test)
show3 = round(acc3 * 100)
print("Resultado modelo 3(em %): ", format(show3))

acc4 = metrics.accuracy_score(result4, y_test)
show4 = round(acc4 * 100)
print("Resultado modelo 4(em %): ", format(show4))
```
