```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/tmoura/machinelearning/master/transfusion.data"
dataset = pd.read_csv(url, header=None)
dataset
```

#NormalizaÃ§Ã£o dos atributos


```python
from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()

dataset[[0]] = mms.fit_transform(dataset[[0]])
dataset[[1]] = mms.fit_transform(dataset[[1]])
dataset[[2]] = mms.fit_transform(dataset[[2]])
dataset[[3]] = mms.fit_transform(dataset[[3]])
dataset[[4]] = mms.fit_transform(dataset[[4]])

dataset
```

#Label Encoder


```python
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()

dataset[0] = label_encoder.fit_transform(dataset[0])
```

```python
dataset
```

#Separar Conjunto de Treino e Teste

```python
y = dataset[0]

X = dataset.drop([0], axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=None,stratify=y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

model1 = DecisionTreeClassifier(criterion="entropy")
model2 = KNeighborsClassifier(n_neighbors=7, metric='euclidean',algorithm='brute')

model3 = DecisionTreeClassifier(criterion="log_loss", max_depth=64, splitter="best")
model4 = KNeighborsClassifier(n_neighbors=5, metric='minkowski',algorithm='ball_tree')


model1.fit(X_train, y_train)
model2.fit(X_train, y_train)
model3.fit(X_train, y_train)
model4.fit(X_train, y_train)
```

```python
from sklearn import metrics

result1 = model1.predict(X_test)
result2 = model2.predict(X_test)
result3 = model3.predict(X_test)
result4 = model4.predict(X_test)

acc1 = metrics.accuracy_score(result1, y_test)
show1 = round(acc1 * 100)
print("Resultado modelo 1(em %): ", format(show1))

acc2 = metrics.accuracy_score(result2, y_test)
show2 = round(acc2 * 100)
print("Resultado modelo 2(em %): ", format(show2))

acc3 = metrics.accuracy_score(result3, y_test)
show3 = round(acc3 * 100)
print("Resultado modelo 3(em %): ", format(show3))

acc4 = metrics.accuracy_score(result4, y_test)
show4 = round(acc4 * 100)
print("Resultado modelo 4(em %): ", format(show4))
```
