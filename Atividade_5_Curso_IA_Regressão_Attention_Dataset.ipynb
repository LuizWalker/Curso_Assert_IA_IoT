```python
## Exercício 3 - Dataset attention do seaborn

# 1) Carregar o dataset "attention" do seaborn
# 2) Fazer um LabelEncoder na coluna "attention"
# 3) Separar o dataframe, deixando a coluna 'attention' para o target (coluna objetivo - y) e as colunas "solutions" e "score" para o X
# 4) Normalizar todo o X usando o mesmo código da questão 1
# 5) Separar os dados em treinamento e teste
# 6) Treinar um modelo de regressão logística
# 7) Treinar uma árvore de decisão
# 8) Treinar um KNN para classificação
# 9) Apresentar os resultados de acurárica em percentual
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn import metrics
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, root_mean_squared_error
from sklearn.preprocessing import MinMaxScaler

from seaborn import load_dataset

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn import preprocessing

from sklearn.linear_model import LinearRegression, LogisticRegression #Regressão logística pode ser utilizada para classificação
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.neighbors import KNeighborsRegressor, KNeighborsClassifier

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
data_base = load_dataset("attention")

data_base.drop(data_base.columns[data_base.columns.str.contains('unnamed', case=False)], axis=1, inplace=True) #Remover colunas "Unnamed"
print(data_base.shape)

data_base
```

```python
data_base = data_base.drop(['subject'],axis=1)
data_base
```

```python
mms = MinMaxScaler()

data_base[['solutions']] = mms.fit_transform(data_base[['solutions']])
data_base[['score']] = mms.fit_transform(data_base[['score']])

data_base
```

```python
label_encoder = preprocessing.LabelEncoder()

data_base['attention'] = label_encoder.fit_transform(data_base['attention']) #Label encoder pra coluna alvo, OnehotEncoder pro restante do dataset

data_base
```

```python
y = data_base['attention']
X = data_base[['solutions','score']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=None, stratify=y)
```

```python
model_logistic = LogisticRegression()

model_logistic = model_logistic.fit(X_train, y_train)

resultados_logistic = model_logistic.predict(X_test)

score_logistic = metrics.accuracy_score(resultados_logistic, y_test)

show_logistic = round(score_logistic * 100)
print("{}%".format(show_logistic))
```

```python
model_tree = DecisionTreeClassifier()

model_tree = model_tree.fit(X_train, y_train)

resultados_tree = model_tree.predict(X_test)

score_tree = metrics.accuracy_score(resultados_tree, y_test)

show_tree = round(score_tree * 100)
print("{}%".format(show_tree))
```

```python
model_knn = KNeighborsClassifier()

model_knn = model_knn.fit(X_train, y_train)

resultados_knn = model_knn.predict(X_test)

score_knn = metrics.accuracy_score(resultados_knn, y_test)

show_knn = round(score_knn * 100)
print("{}%".format(show_knn))
```
