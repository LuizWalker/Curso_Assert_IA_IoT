```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn import metrics
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, root_mean_squared_error
from sklearn.preprocessing import MinMaxScaler

from seaborn import load_dataset

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn import preprocessing

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
data_base = load_dataset("car_crashes")
print(data_base.shape)
data_base
```

```python
data_base = data_base.drop(columns=['abbrev'])
data_base
```

```python
mms = MinMaxScaler()

data_base[['speeding']] = mms.fit_transform(data_base[['speeding']])
data_base[['alcohol']] = mms.fit_transform(data_base[['alcohol']])
data_base[['not_distracted']] = mms.fit_transform(data_base[['not_distracted']])
data_base[['no_previous']] = mms.fit_transform(data_base[['no_previous']])
data_base[['ins_premium']] = mms.fit_transform(data_base[['ins_premium']])
data_base[['ins_losses']] = mms.fit_transform(data_base[['ins_losses']])

data_base
```

```python
y = data_base['total']

X = data_base.drop('total', axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=None)
```

```python
Model_Linear = LinearRegression()
Model_Linear.fit(X_train,y_train)

Prediction_Linear = Model_Linear.predict(X_test)
print(mean_squared_error(y_test,Prediction_Linear))
print(mean_absolute_error(y_test,Prediction_Linear))
print(root_mean_squared_error(y_test,Prediction_Linear))
```

```python
Model_tree = DecisionTreeRegressor()
Model_tree.fit(X_train,y_train)

Prediction_tree = Model_tree.predict(X_test)
print(mean_squared_error(y_test,Prediction_tree))
print(mean_absolute_error(y_test,Prediction_tree))
print(root_mean_squared_error(y_test,Prediction_tree))
```

```python
Model_KNN = KNeighborsRegressor()
Model_KNN.fit(X_train,y_train)

Prediction_KNN = Model_KNN.predict(X_test)
print(mean_squared_error(y_test,Prediction_KNN))
print(mean_absolute_error(y_test,Prediction_KNN))
print(root_mean_squared_error(y_test,Prediction_KNN))
```
