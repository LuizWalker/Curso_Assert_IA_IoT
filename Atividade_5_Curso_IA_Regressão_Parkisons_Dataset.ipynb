```python
## Exercício 2 - Dataset Parkinsons (dataset com 2 classes)

# 1) Carregar o dataset "parkinsons" pela URL abaixo:

# url = "https://raw.githubusercontent.com/tmoura/machinelearning/master/parkinsons.data"

# 2) Separar os dados y e X. A coluna y é a coluna 0 (zero)
# 3) Normalizar todas as colunas de X (usando o mesmo código da questão anterior)
# 4) Separar oa dados em treinamento e teste
# 5) Treinar um modelo de regressão logística
# 6) Treinar uma árvore de decisão
# 7) Treinar um KNN para classificação
# 8) Apresentar os resultados de acurácia em percentual
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn import metrics
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, root_mean_squared_error
from sklearn.preprocessing import MinMaxScaler

from seaborn import load_dataset

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn import preprocessing

from sklearn.linear_model import LinearRegression, LogisticRegression #Regressão logística pode ser utilizada para classificação
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.neighbors import KNeighborsRegressor, KNeighborsClassifier

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
url = "https://raw.githubusercontent.com/tmoura/machinelearning/master/parkinsons.data"

data_base = pd.read_csv(url,header=None)

data_base

data_base.shape
```

```python
mms = MinMaxScaler()

for i in range(1,23):
  data_base[[i]] = mms.fit_transform(data_base[[i]])


data_base
```

```python
y = data_base[0]
X = data_base.drop(0,axis=1)

X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=None, stratify=y)
```

```python
model_logistic = LogisticRegression()

model_logistic = model_logistic.fit(X_train, y_train)

resultados_logistic = model_logistic.predict(X_test)

score_logistic = metrics.accuracy_score(resultados_logistic, y_test)

show_logistic = round(score_logistic * 100)
print("{}%".format(show_logistic))
```

```python
model_tree = DecisionTreeClassifier()

model_tree = model_tree.fit(X_train, y_train)

resultados_tree = model_tree.predict(X_test)

score_tree = metrics.accuracy_score(resultados_tree, y_test)

show_tree = round(score_tree * 100)
print("{}%".format(show_tree))
```

```python
model_knn = KNeighborsClassifier()

model_knn = model_knn.fit(X_train, y_train)

resultados_knn = model_knn.predict(X_test)

score_knn = metrics.accuracy_score(resultados_knn, y_test)

show_knn = round(score_knn * 100)
print("{}%".format(show_knn))
```
